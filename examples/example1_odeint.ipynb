\begin{cases}
\dot{y_1} = y_2,
\\
\dot{y_2} = - b y_1 - c \sin(y_2)
\end{cases}

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def sys_of_funcs(y, t, b, c):
    y1 = y[0]
    y2 = y[1]
    f1 = y2
    f2 = - b * y1 - c * np.sin(y2) 
    return [f1, f2]
```

```python
init = np.pi - 0.1, 0.0        # (3.041592653589793, 0.0)
t = np.linspace(0, 10, 101)
b = 1
c = -1
sol = odeint(sys_of_funcs, init, t, args=(b, c))
```

```python
plt.grid()
plt.xlabel('t')
plt.ylabel('y')
plt.plot(t, sol[:, 0], color='b', label=r"$y_1$")
plt.plot(t, sol[:, 1], color='c', label=r"$y_2$")
plt.legend(loc="best")
plt.show()
```
